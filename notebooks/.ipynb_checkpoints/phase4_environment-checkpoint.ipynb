```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.animation as animation
from matplotlib.patches import FancyArrowPatch
import gymnasium as gym
from gymnasium import spaces
import torch
import torch.nn as nn
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
```

```python
# Load master dataset
df = pd.read_csv('../data/master_dataset.csv')
df['datetime'] = pd.to_datetime(df['datetime'])

# Load scalers (recreate from Phase 3)
from sklearn.preprocessing import MinMaxScaler

input_features = ['ssrd_wm2', 'tp', 'temp_c', 'load_kw',
                  'location_code', 'hour', 'month', 'dayofweek']

df['location_code'] = df['location'].map({'Tamale': 0, 'Kumasi': 1, 'Axim': 2})
df['hour']          = df['datetime'].dt.hour
df['month']         = df['datetime'].dt.month
df['dayofweek']     = df['datetime'].dt.dayofweek

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
scaler_X.fit(df[input_features])
scaler_y.fit(df[['ssrd_wm2', 'load_kw']])

# Recreate LSTM model
class MiniGridLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, forecast, output_size):
        super(MiniGridLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers  = num_layers
        self.forecast    = forecast
        self.output_size = output_size
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers,
                            batch_first=True, dropout=0.2)
        self.fc   = nn.Linear(hidden_size, forecast * output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out.view(-1, self.forecast, self.output_size)

# Load trained LSTM
lstm_model = MiniGridLSTM(8, 128, 2, 24, 2)
lstm_model.load_state_dict(torch.load('../models/best_lstm.pth'))
lstm_model.eval()

print("✅ Data and LSTM model loaded!")
print(f"Dataset shape: {df.shape}")
```

```python
class MiniGridEnv(gym.Env):
    def __init__(self, df, lstm_model, scaler_X, scaler_y, mode='train'):
        super(MiniGridEnv, self).__init__()

        self.df        = df.reset_index(drop=True)
        self.lstm      = lstm_model
        self.scaler_X  = scaler_X
        self.scaler_y  = scaler_y
        self.mode      = mode

        # ── Mini-grid specs ──────────────────────────────────────
        self.battery_capacity  = 200.0   # kWh
        self.max_charge_rate   = 50.0    # kW
        self.max_discharge_rate= 50.0    # kW
        self.charge_efficiency = 0.95
        self.discharge_efficiency = 0.95
        self.soc_min           = 0.10    # 10% minimum SOC
        self.soc_max           = 0.95    # 95% maximum SOC
        self.initial_soh       = 1.0     # 100% health
        self.soh_deg_per_cycle = 0.0001  # degradation per cycle

        # ── Action space: continuous [-1, 1] ─────────────────────
        # -1 = full discharge, 0 = idle, +1 = full charge
        self.action_space = spaces.Box(
            low=-1.0, high=1.0, shape=(1,), dtype=np.float32)

        # ── Observation space ────────────────────────────────────
        # [SOC, SOH, solar_forecast(24), load_forecast(24), hour, month]
        self.observation_space = spaces.Box(
            low=-np.inf, high=np.inf, shape=(52,), dtype=np.float32)

        self.reset()

    # ── Reset ────────────────────────────────────────────────────
    def reset(self, seed=None):
        self.soc      = 0.5   # start at 50% charge
        self.soh      = self.initial_soh
        self.rul      = 1.0
        self.efc      = 0.0
        self.ens      = 0.0   # energy not served
        self.step_idx = np.random.randint(24, len(self.df) - 24 - 48)
        return self._get_obs(), {}

    # ── Get LSTM forecast ────────────────────────────────────────
    def _get_forecast(self):
        row = self.df.iloc[self.step_idx]
        lookback = self.df.iloc[self.step_idx - 24 : self.step_idx]
        X = self.scaler_X.transform(lookback[
            ['ssrd_wm2','tp','temp_c','load_kw',
             'location_code','hour','month','dayofweek']])
        X_tensor = torch.FloatTensor(X).unsqueeze(0)
        with torch.no_grad():
            forecast = self.lstm(X_tensor).numpy()[0]  # (24, 2)
        forecast_inv = self.scaler_y.inverse_transform(
            forecast.reshape(-1, 2))
        solar_fc = np.clip(forecast_inv[:, 0], 0, None)
        load_fc  = np.clip(forecast_inv[:, 1], 0, None)
        return solar_fc, load_fc

    # ── Observation ──────────────────────────────────────────────
    def _get_obs(self):
        solar_fc, load_fc = self._get_forecast()
        row   = self.df.iloc[self.step_idx]
        obs   = np.concatenate([
            [self.soc, self.soh],
            solar_fc / 1000.0,   # normalize
            load_fc  / 50.0,     # normalize
            [row['hour'] / 23.0, row['month'] / 12.0]
        ]).astype(np.float32)
        return obs

    # ── Step ─────────────────────────────────────────────────────
    def step(self, action):
        action     = float(np.clip(action[0], -1.0, 1.0))
        row        = self.df.iloc[self.step_idx]
        solar_kw   = row['ssrd_wm2'] * 0.05   # 50m² panel @ ~5% effective
        load_kw    = row['load_kw']

        # Battery charge/discharge
        if action > 0:   # charging
            charge_kw  = action * self.max_charge_rate
            charge_kw  = min(charge_kw,
                             (self.soc_max - self.soc) * self.battery_capacity)
            energy_in  = charge_kw * self.charge_efficiency
            self.soc  += energy_in / self.battery_capacity
            net_load   = load_kw + charge_kw - solar_kw
        else:            # discharging
            discharge_kw = abs(action) * self.max_discharge_rate
            discharge_kw = min(discharge_kw,
                               (self.soc - self.soc_min) * self.battery_capacity)
            energy_out   = discharge_kw * self.discharge_efficiency
            self.soc    -= discharge_kw / self.battery_capacity
            net_load     = load_kw - solar_kw - energy_out

        self.soc = np.clip(self.soc, self.soc_min, self.soc_max)

        # Energy not served
        ens_step = max(0.0, net_load)
        self.ens += ens_step

        # Battery degradation
        self.efc     += abs(action) * 0.5
        self.soh      = max(0.0, self.initial_soh -
                            self.efc * self.soh_deg_per_cycle)
        self.rul      = self.soh / self.initial_soh

        # ── Reward ───────────────────────────────────────────────
        reward  =  1.0                          # base reward per step
        reward -= ens_step        * 2.0         # penalize unserved energy
        reward -= abs(action)     * 0.1         # penalize battery stress
        reward += (self.soh - 0.8)* 5.0         # reward good battery health
        reward -= (self.soc < 0.2)* 2.0         # penalize low SOC
        reward -= (self.soc > 0.9)* 0.5         # penalize overcharging

        self.step_idx += 1
        done = self.step_idx >= len(self.df) - 25

        return self._get_obs(), reward, done, False, {
            'soc': self.soc, 'soh': self.soh,
            'rul': self.rul, 'ens': ens_step,
            'solar_kw': solar_kw, 'load_kw': load_kw,
            'action': action
        }

print("✅ MiniGridEnv class defined!")
```

```python
# Instantiate and test the environment
env = MiniGridEnv(df, lstm_model, scaler_X, scaler_y)

# Run one test episode
obs, _ = env.reset()
print("✅ Environment created!")
print(f"Observation space: {env.observation_space.shape}")
print(f"Action space: {env.action_space.shape}")
print(f"Initial SOC: {env.soc:.2f}")
print(f"Initial SOH: {env.soh:.2f}")
print(f"Initial observation shape: {obs.shape}")

# Take 5 random steps to test
print("\n--- Testing 5 random steps ---")
for i in range(5):
    action = env.action_space.sample()
    obs, reward, done, _, info = env.step(action)
    print(f"Step {i+1} | Action: {info['action']:+.2f} | "
          f"SOC: {info['soc']:.3f} | SOH: {info['soh']:.4f} | "
          f"Solar: {info['solar_kw']:.2f}kW | Load: {info['load_kw']:.2f}kW | "
          f"ENS: {info['ens']:.3f} | Reward: {reward:.3f}")

print("\n✅ Environment working correctly!")
```

```python
def rule_based_controller(soc, solar_kw, load_kw, 
                           soc_min=0.10, soc_max=0.95):
    """
    Simple rule-based controller:
    - If solar > load → charge battery with excess
    - If solar < load → discharge battery to cover deficit
    - Respect SOC limits
    """
    net_power = solar_kw - load_kw  # positive = excess solar

    if net_power > 0:
        # Excess solar — charge battery
        if soc < soc_max:
            action = min(net_power / 50.0, 1.0)   # normalize to [0,1]
        else:
            action = 0.0  # battery full
    else:
        # Deficit — discharge battery
        if soc > soc_min:
            action = max(net_power / 50.0, -1.0)  # normalize to [-1,0]
        else:
            action = 0.0  # battery empty

    return np.array([action], dtype=np.float32)


# Test rule-based controller for one full episode
env_rb = MiniGridEnv(df, lstm_model, scaler_X, scaler_y)
obs, _ = env_rb.reset()

rb_soc, rb_soh, rb_ens, rb_solar, rb_load, rb_actions = [], [], [], [], [], []

done = False
total_reward = 0
while not done:
    action = rule_based_controller(
        env_rb.soc,
        env_rb.df.iloc[env_rb.step_idx]['ssrd_wm2'] * 0.05,
        env_rb.df.iloc[env_rb.step_idx]['load_kw']
    )
    obs, reward, done, _, info = env_rb.step(action)
    total_reward += reward
    rb_soc.append(info['soc'])
    rb_soh.append(info['soh'])
    rb_ens.append(info['ens'])
    rb_solar.append(info['solar_kw'])
    rb_load.append(info['load_kw'])
    rb_actions.append(info['action'])

print("✅ Rule-Based Controller tested!")
print(f"Total Steps:     {len(rb_soc)}")
print(f"Total Reward:    {total_reward:.2f}")
print(f"Final SOC:       {rb_soc[-1]:.3f}")
print(f"Final SOH:       {rb_soh[-1]:.4f}")
print(f"Total ENS:       {sum(rb_ens):.2f} kWh")
print(f"Avg SOC:         {np.mean(rb_soc):.3f}")
```
